1) Lists (mutable)

```python
# List example (mutable)
my_list = ["AQI", "PM2.5", "O3"]

print("Original list:", my_list)
print("Index access [0]:", my_list[0])

my_list.append("NO2")
print("After append:", my_list)

my_list[1] = "PM10"   # modify element by index
print("After modifying index 1:", my_list)

removed = my_list.pop()  # remove last item
print("Removed:", removed)
print("Final list:", my_list)
```

2) Tuples (immutable)

```python
# Tuple example (immutable)
my_tuple = ("Chennai", "Moderate", 120)

print("Tuple:", my_tuple)
print("Index access [0]:", my_tuple[0])

try:
    my_tuple[0] = "Bengaluru"  # should fail (tuples are immutable)
except TypeError as e:
    print("TypeError (expected):", e)
```

3) Dictionaries (key-value)

```python
# Dictionary example (key-value)
person = {"name": "Dhany", "age": 19}

print("Original dictionary:", person)
print("Value by key 'name':", person["name"])

person["age"] = 20          # modify existing key
person["city"] = "Chennai" # add new key
print("Updated dictionary:", person)

# Optional: show all keys
print("Keys:", list(person.keys()))
```
