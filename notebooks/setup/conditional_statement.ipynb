# Conditional Statements (if / elif / else)
This notebook demonstrates decision-making in Python using `if`, `if-else`, `if-elif-else`, and logical operators.

## 1) Simple `if`
This code prints only when the condition is true. When the condition is false, nothing inside the block runs.

```python
scores = [30, 85]

for s in scores:
    if s >= 80:
        print(f"Score {s}: qualifies (>= 80)")
```

## 2) `if-else`
This code handles both outcomes: one message for true, another for false.

```python
temperature = 28

if temperature >= 30:
    print("Hot day (>= 30°C)")
else:
    print("Not too hot (< 30°C)")
```

## 3) `if-elif-else`
This chooses exactly one branch based on multiple conditions.

```python
score = 67

if score >= 75:
    print("Grade: A")
elif score >= 50:
    print("Grade: B")
else:
    print("Grade: C")
```

```python
## 4) Logical operators (`and`, `or`, `not`)
- `and`: both conditions must be true
- `or`: at least one condition must be true
- `not`: flips a boolean value
```

```python
temp = 30
humidity = 70
is_raining = False

is_hot = temp >= 30
is_humid = humidity >= 60

if is_hot and is_humid:
    print("Condition: both hot AND humid")

if is_hot or is_humid:
    print("Condition: hot OR humid (at least one is true)")

if not is_raining:
    print("Condition: NOT raining (dry conditions)")
```

Summary:
- `if` runs only when the condition is true.
- `if-else` handles both true/false cases.
- `if-elif-else` picks one path among many.
- Logical operators combine conditions safely.
