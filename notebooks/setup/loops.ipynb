# For and While Loops

This notebook shows `for` loops, `while` loops, and loop control with `break` / `continue`.

## 1) `for` loop over `range`

Repeats a fixed number of times. Good when you know how many steps you need.

```python
print("for loop with range(1, 6):")
for i in range(1, 6):
    print(f"  i = {i}, i squared = {i * i}")
```

## 2) `for` loop over a list

Visits each item in order. Good for collections (lists, tuples, etc.).

```python
pollutants = ["PM2.5", "O3", "NO2"]

print("for loop over a list:")
for name in pollutants:
    print(f"  processing: {name}")
```

## 3) `while` loop (condition-based)

Repeats while a condition is true. You must **update** variables so the condition can become false.

```python
count = 0

print("while loop (stops when count reaches 3):")
while count < 3:
    print(f"  count = {count}")
    count += 1

print("done; count is now:", count)
```

## 4) `continue` — skip an iteration

Use when you want to skip the rest of the current loop body and go to the next iteration.

```python
print("for i in range(1, 6), skip even numbers with continue:")
for i in range(1, 6):
    if i % 2 == 0:
        continue
    print(f"  odd only: i = {i}")
```

## 5) `break` — exit the loop early

Use when you want to stop the whole loop as soon as a condition is met.

```python
values = [3, 7, 12, 4, 9]

print("find first value > 10, then break:")
for v in values:
    print(f"  checking {v}")
    if v > 10:
        print(f"  found {v} — stopping loop")
        break
```

## 6) Infinite loops (what to avoid)

A `while` loop runs forever if the condition **never** becomes false.

Common causes:
- Forgetting to update the variable tested in the condition
- Writing `while True:` without a `break` or a condition that eventually stops

**Summary**
- `for`: iterate over a sequence or `range`.
- `while`: repeat until a condition is false; update state each iteration.
- `continue`: skip to next iteration.
- `break`: leave the loop immediately.
