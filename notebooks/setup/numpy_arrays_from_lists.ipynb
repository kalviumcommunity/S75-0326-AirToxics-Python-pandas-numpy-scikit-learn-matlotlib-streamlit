# NumPy Arrays from Python Lists

This notebook demonstrates how to:
- Convert Python lists into NumPy arrays (1D and 2D)
- Inspect `shape`, `ndim`, and `dtype`
- Do basic arithmetic with element-wise behavior
- Compare list behavior vs NumPy array behavior

## 1) Import NumPy

```python
import numpy as np
print("NumPy version:", np.__version__)
```

## 2) Create a 1D array from a Python list

```python
py_list_1d = [1, 2, 3, 4]

arr_1d = np.array(py_list_1d)

print("Array:", arr_1d)
print("shape:", arr_1d.shape)
print("ndim:", arr_1d.ndim)
print("dtype:", arr_1d.dtype)
```

## 3) Create a 2D array from a nested Python list

```python
py_list_2d = [[1, 2],
              [3, 4],
              [5, 6]]

arr_2d = np.array(py_list_2d)

print("Array:\n", arr_2d)
print("shape:", arr_2d.shape)
print("ndim:", arr_2d.ndim)
print("dtype:", arr_2d.dtype)
```

## 4) Basic operations (element-wise)

```python
arr = np.array([10, 20, 30])

print("arr:", arr)
print("arr + 5:", arr + 5)
print("arr * 2:", arr * 2)

other = np.array([1, 2, 3])
print("arr + other:", arr + other)
```

## 5) Lists vs NumPy arrays (key difference)

Python lists support repetition with `*`.
NumPy arrays do element-wise numeric multiplication with `*`.

```python
py_list = [1, 2, 3]
np_arr = np.array([1, 2, 3])

print("py_list * 2:", py_list * 2)
print("np_arr * 2:", np_arr * 2)
```

## Summary
- `np.array()` converts lists into NumPy arrays.
- Arrays have `shape`, `ndim`, and `dtype` you can inspect.
- Arithmetic on NumPy arrays is element-wise (vectorized).
- List `*` repeats; NumPy `*` multiplies values.
