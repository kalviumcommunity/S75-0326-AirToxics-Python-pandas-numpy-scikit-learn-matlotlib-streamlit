# Code Structure for Readability and Reuse

This notebook demonstrates a clean Python script layout:
- Imports and configuration first
- Reusable helper functions
- Minimal top-level execution (via `main()`)

## What you’ll see
We’ll calculate an average from repeated AQI readings, classify it into a label, and generate a report for multiple cities.

```python
# ----------------------------
# Imports (top of file)
# ----------------------------
from typing import List, Dict

# ----------------------------
# Configuration / data (setup)
# ----------------------------
aqi_thresholds = {
    "Good": 100,
    "Moderate": 200,  # anything <= 200 after that is "Moderate"
    "Poor": float("inf"),
}

aqi_data_by_city: Dict[str, List[float]] = {
    "Chennai": [120, 110, 90],
    "Delhi": [180, 210, 160],
    "Mumbai": [70, 85, 95],
}
```

```python
# ----------------------------
# Helper functions (logic)
# ----------------------------
def calculate_average(values: List[float]) -> float:
    """Return the mean of a list of numbers."""
    if not values:
        raise ValueError("values must not be empty")
    return sum(values) / len(values)


def classify_aqi(average_aqi: float) -> str:
    """Classify AQI using simple thresholds for this demo."""
    if average_aqi <= aqi_thresholds["Good"]:
        return "Good"
    if average_aqi <= aqi_thresholds["Moderate"]:
        return "Moderate"
    return "Poor"


def build_city_report(city: str, readings: List[float]) -> str:
    """Create a readable report line for one city."""
    avg = calculate_average(readings)
    category = classify_aqi(avg)
    return f"{city}: average AQI = {avg:.1f} -> {category}"
```

## Execution (minimal + readable)

The `main()` function is the only place we run the overall workflow.

```python
def main() -> None:
    reports: List[str] = []

    # Reuse `build_city_report()` for multiple inputs
    for city, readings in aqi_data_by_city.items():
        reports.append(build_city_report(city, readings))

    print("City Reports:")
    for line in reports:
        print("-", line)


if __name__ == "__main__":
    main()
```

Summary:
- Imports/config at the top
- Reusable helper functions for calculation + classification + report creation
- Clean execution via `main()`
