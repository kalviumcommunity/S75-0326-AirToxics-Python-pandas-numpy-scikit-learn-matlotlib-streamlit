# Creating Pandas DataFrames from Dictionaries and Files

This milestone focuses on creating Pandas DataFrames from dictionaries and files, which is a core skill for working with real-world data. DataFrames are the primary structure used in Pandas to represent tabular data similar to spreadsheets or database tables.

## Learning Objectives

- Understand what a Pandas DataFrame represents
- Create DataFrames from Python dictionaries
- Load DataFrames from common file formats
- Inspect DataFrame structure and contents
- Recognize common issues during data loading

## 1. Understanding Pandas DataFrames

A Pandas DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns). It's similar to a spreadsheet or a SQL table.

### Key Concepts:
- **Rows**: Horizontal entries, each representing a single record or observation
- **Columns**: Vertical entries, each representing a variable or feature
- **Column Names**: Labels for each column
- **Index**: Labels for each row (default is 0, 1, 2, ...)

DataFrames are the primary data structure for data manipulation and analysis in Pandas.

## 2. Creating DataFrames from Dictionaries

You can create a DataFrame from a Python dictionary where:
- Keys become column names
- Values become the column data (as lists or arrays)

Let's create a simple example with air quality data.

```python
import pandas as pd

# Create a dictionary with air quality data
air_quality_data = {
    'City': ['Chennai', 'Delhi', 'Mumbai', 'Kolkata'],
    'AQI': [120, 200, 150, 180],
    'PM2.5': [45.2, 85.6, 62.1, 78.3]
}

# Convert dictionary to DataFrame
df_from_dict = pd.DataFrame(air_quality_data)

# Display the DataFrame
print("DataFrame created from dictionary:")
print(df_from_dict)
print("\nColumn names:", list(df_from_dict.columns))
print("Index:", list(df_from_dict.index))
```

## 3. Creating DataFrames from Files

Pandas can load data from various file formats including CSV, Excel, JSON, etc. The most common is CSV (Comma-Separated Values).

Key parameters for `pd.read_csv()`:
- `filepath`: Path to the file
- `header`: Row number to use as column names (default 0)
- `index_col`: Column to use as row labels
- `sep`: Delimiter (default ',')
- `encoding`: File encoding

Let's load data from the sample CSV file.

```python
# Load DataFrame from CSV file
csv_file_path = '../data/raw/sample_raw_data.csv'  # Relative path from notebooks/setup/
df_from_csv = pd.read_csv(csv_file_path)

# Display the loaded DataFrame
print("DataFrame loaded from CSV:")
print(df_from_csv)
print("\nShape:", df_from_csv.shape)
print("Columns:", list(df_from_csv.columns))
```

## 4. Inspecting DataFrame Structure

After loading data, it's crucial to inspect the DataFrame to understand its structure and contents. This helps identify any issues early.

Common inspection methods:
- `head()`: View first few rows
- `tail()`: View last few rows
- `info()`: Summary of columns, data types, and non-null values
- `describe()`: Statistical summary for numeric columns
- `shape`: Dimensions (rows, columns)
- `columns`: Column names
- `dtypes`: Data types of each column

```python
# Inspect the DataFrame from dictionary
print("=== Inspection of DataFrame from Dictionary ===")
print("First 3 rows:")
print(df_from_dict.head(3))
print("\nDataFrame info:")
print(df_from_dict.info())
print("\nStatistical summary:")
print(df_from_dict.describe())
print("\nShape:", df_from_dict.shape)
print("Data types:")
print(df_from_dict.dtypes)
```

```python
# Inspect the DataFrame from CSV
print("=== Inspection of DataFrame from CSV ===")
print("First 3 rows:")
print(df_from_csv.head(3))
print("\nDataFrame info:")
print(df_from_csv.info())
print("\nShape:", df_from_csv.shape)
print("Data types:")
print(df_from_csv.dtypes)
```

## Summary

In this milestone, you have learned to:

1. **Understand DataFrames**: Two-dimensional tabular data structure with rows and columns
2. **Create from Dictionaries**: Use `pd.DataFrame()` with a dictionary where keys are columns
3. **Load from Files**: Use `pd.read_csv()` and similar functions for file-based data
4. **Inspect Structure**: Use methods like `head()`, `info()`, `describe()`, and attributes like `shape`, `columns`, `dtypes`

These skills form the foundation for all data analysis work in Pandas. Remember to always inspect your data after loading to ensure it's structured correctly!

# Loading CSV Data into Pandas DataFrames

This milestone focuses on loading CSV data into Pandas DataFrames, which is one of the most common tasks in real-world Data Science workflows. CSV files are a standard format for sharing tabular data, and knowing how to load them correctly is essential before any analysis can begin.

## Learning Objectives

- Understand what CSV files represent
- Load CSV files into Pandas DataFrames
- Interpret headers, rows, and columns correctly
- Inspect loaded data for correctness
- Recognize common CSV loading issues

## 1. Understanding CSV Files

CSV (Comma-Separated Values) files are a plain text format for storing tabular data. Each line represents a row, and values within each row are separated by commas (or other delimiters).

### Key Concepts:
- **Headers**: The first row usually contains column names
- **Data Rows**: Subsequent rows contain the actual data
- **Delimiters**: Commas separate values (other delimiters like tabs or semicolons can be used)
- **Structure**: Rectangular format with consistent columns across all rows

CSV files are like simplified spreadsheets saved as text.

```python
# Let's first examine the raw CSV file content
print("Raw CSV file content:")
with open('../data/raw/sample_raw_data.csv', 'r') as f:
    content = f.read()
    print(content)
print("\nNotice:")
print("- First line contains headers: 'city,AQI'")
print("- Second line contains data: 'Chennai,120'")
print("- Third line contains data: 'Delhi,200'")
print("- Comma separates values within each row")
```

## 2. Loading CSV Files into Pandas

Pandas provides the `pd.read_csv()` function to load CSV files into DataFrames. This function automatically handles headers, data types, and creates the appropriate DataFrame structure.

### Basic Loading Process:
1. Provide the file path
2. Pandas reads the file
3. First row becomes column names (by default)
4. Remaining rows become data
5. Data types are inferred automatically

```python
# Load the CSV file into a DataFrame
csv_df = pd.read_csv('../data/raw/sample_raw_data.csv')

print("DataFrame loaded from CSV:")
print(csv_df)
print("\nDataFrame type:", type(csv_df))
print("Shape (rows, columns):", csv_df.shape)
```

## 3. Inspecting Loaded Data

After loading CSV data, always inspect the DataFrame to ensure it was loaded correctly. This helps catch issues like incorrect column names, missing data, or unexpected data types.

### Key Inspection Steps:
1. Check the shape (dimensions)
2. View column names
3. Preview the first few rows
4. Check data types
5. Look for any obvious issues

```python
# Inspect the loaded DataFrame
print("=== DataFrame Inspection ===")
print("Shape:", csv_df.shape)
print("Columns:", list(csv_df.columns))
print("Data types:")
print(csv_df.dtypes)
print("\nFirst 5 rows:")
print(csv_df.head())
print("\nDataFrame info:")
print(csv_df.info())
```

## 4. Recognizing Common Loading Issues

CSV loading can sometimes go wrong. Here are some common issues and how to recognize them:

### Common Issues:
1. **Wrong delimiter**: Using comma when file uses semicolons
2. **No header**: Data treated as column names when it shouldn't be
3. **Encoding issues**: Special characters not displayed correctly
4. **Missing values**: Empty cells not handled properly
5. **Incorrect data types**: Numbers loaded as strings

### Prevention:
- Always inspect the raw file first
- Check column names after loading
- Verify data types are correct
- Look for unexpected missing values

```python
# Example of a common loading issue: loading without header
print("=== Example: Loading CSV without header ===")
df_no_header = pd.read_csv('../data/raw/sample_raw_data.csv', header=None)
print("DataFrame loaded WITHOUT header:")
print(df_no_header)
print("\nColumns:", list(df_no_header.columns))
print("Notice how 'city' and 'AQI' became data instead of column names!")

print("\n=== Correct loading with header ===")
df_with_header = pd.read_csv('../data/raw/sample_raw_data.csv', header=0)  # header=0 is default
print("DataFrame loaded WITH header:")
print(df_with_header)
print("Columns:", list(df_with_header.columns))
```
